```python
#1
def unique_sorted_student_ids(list1, list2):
    return sorted(set(list1) | set(list2))
list1 = input("Nhập list ID thứ nhất: ").split()
list2 = input("Nhập list ID thứ hai: ").split()

result = unique_sorted_student_ids(list1, list2)
print(result)
```

```python
#2
def top3_frequent_names(names):
    freq = {}
    for name in names:
        freq[name] = freq.get(name, 0) + 1

        sorted_names = sorted(freq.items(), key=lambda x: x[1], reverse=True)


    result = []
    for i in range(min(3, len(sorted_names))):
        result.append(sorted_names[i][0])

    return result



names = input("Nhập danh sách tên: ").split()
print(top3_frequent_names(names))
```

```python
#3
def most_frequent_triple(triples):
    if not triples:
        raise ValueError

    freq = {}
    for t in triples:
        freq[t] = freq.get(t, 0) + 1

    return max(freq, key=freq.get)


triples = eval(input())
print(most_frequent_triple(triples))
```

```python
#4
def reconcile_users(registered, paid):
    set_r = set(registered)
    set_p = set(paid)

    paid_not_registered = sorted(set_p - set_r)
    registered_not_paid = sorted(set_r - set_p)
    both = sorted(set_r & set_p)

    return paid_not_registered, registered_not_paid, both


registered = input().split()
paid = input().split()

print(reconcile_users(registered, paid))
```

```python
#5
def contains_duplicate(nums):
    return len(nums) != len(set(nums))


nums = list(map(int, input().split()))
print(contains_duplicate(nums))
```

```python
#6
def squares_div_by3_not6(arr):
    return [x * x for x in arr if x % 3 == 0 and x % 6 != 0]


arr = list(map(int, input().split()))
print(squares_div_by3_not6(arr))
```

```python
#7
def even_numbers_in_matrix(matrix):
    return [x for row in matrix for x in row if x % 2 == 0]


matrix = eval(input())
print(even_numbers_in_matrix(matrix))
```

```python
#8
def flatten_2d(matrix):
    return [x for row in matrix for x in row]


matrix = eval(input())
print(flatten_2d(matrix))
```

```python
#9
def contains_nearby_duplicate(nums, k):
    last_index = {}

    for i, num in enumerate(nums):
        if num in last_index and i - last_index[num] <= k:
            return True
        last_index[num] = i

    return False


nums = list(map(int, input().split()))
k = int(input())
print(contains_nearby_duplicate(nums, k))
```

```python
#10
def merge_similar_items(items1, items2):
    mp = {}

    for v, w in items1:
        mp[v] = mp.get(v, 0) + w

    for v, w in items2:
        mp[v] = mp.get(v, 0) + w

    return [[v, mp[v]] for v in sorted(mp)]


items1 = eval(input())
items2 = eval(input())
print(merge_similar_items(items1, items2))
```

```python
#11
def farm_configurations(n):
    result = []
    for cows in range(n // 4 + 1):
        remaining = n - 4 * cows
        if remaining % 2 == 0:
            chickens = remaining // 2
            result.append((chickens, cows))
    return result


n = int(input())
print(farm_configurations(n))
```

```python
#12
def can_make_odd_sum(a):
    odd = sum(x % 2 for x in a)
    even = len(a) - odd
    return odd > 0 and even > 0


a = list(map(int, input().split()))
print(can_make_odd_sum(a))

```

```python
#13
def balanced_array(n):
    if (n // 2) % 2 != 0:
        return ("NO", [])

    half = n // 2
    evens = [2 * i for i in range(1, half + 1)]
    odds = [2 * i - 1 for i in range(1, half)]

    last_odd = sum(evens) - sum(odds)
    odds.append(last_odd)

    return ("YES", evens + odds)


n = int(input())
print(balanced_array(n))
```
