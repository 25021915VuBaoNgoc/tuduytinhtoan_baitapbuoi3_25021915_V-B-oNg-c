```python
#W8A3
def happy_num(n):
    seen = set()
    while n != 1 and n not in seen:
        seen.add(n)
        n = sum(int(digit)**2 for digit in str(n))
    return n == 1
try:
    a = int(input())
    if happy_num(a):
        print('YES')
    else:
        print('NO')

except ValueError:
    print('Nhập vào 1 số nguyên dương')
```

```python
#W8A11
import re

try:
    t = int(input())
    
    for _ in range(t):
        s = input().strip()
        try:
            re.compile(s)
            print("Valid")
        except re.error:
            print("Invalid")
except EOFError:
    pass
```
