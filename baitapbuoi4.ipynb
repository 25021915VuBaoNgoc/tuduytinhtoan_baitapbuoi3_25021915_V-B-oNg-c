```python
#W4A2
from math import*
def nt(n):
  if n <=0:
    print('Nhập lại số dương')
  else:
    if n == 2:
      return True
    for i in range(2,isqrt(n)+1):
      if n % i ==0:
        print(f"{n} không là số nguyên tố")
        return False
    print(f"{n} là số nguyên tố")
    return True
if __name__ =="__main__":
  n=int(input())
  nt(n)
```

```python
#W4A6
from math import*
def nt(n):
  if n==2:
    return True
  for i in range(2,isqrt(n)+1):
    if n %i==0:
      return False
  return True
def tong(a,b):
  sum=0
  for i in range(a,b+1):
    if nt(i):
      sum+=i
  return sum
if __name__=="__main__":
  tc=int(input())
  for i in range(1,tc+1):
    a,b=map(int,input().split())
    print(tong(a,b),"\n")
```

```python
#W4A7
from math import*
def nt_max(n):
  if n==2:
    print(2)
    return True
  for i in range(2,isqrt(n)+1):
    while n%i==0:
      uoc_max=i
      n//=i
  if n>1:
    uoc_max=n
  return uoc_max
if __name__=="__main__":
  n=int(input())
  print(nt_max(n))
```

```python
#W4A8
def rev(n):
  new=0
  while n!=0:
    new=new*10+n%10
    n//=10
  return new
if __name__=="__main__":
  n=int(input("Nhập n :"))
  step=0
  while n != rev(n):
    n = n + rev(n)
    step += 1
  print(f"số đối xứng thu đc là {n}")
  print(f"Số bước {step}")
```

```python
#W4A10
def collatz(n):
  length=1
  while n!=1:
    if n%2==0:
      n//=2
    else:
      n=3*n+1
    length+=1
  return length
if __name__=="__main__":
  n=int(input())
  max_len=0
  best_n=1
  for i in range(1, n + 1):
        L = collatz(i)
        if L > max_len or (L == max_len and i < best_n):
            max_len = L
            best_n = i
  print(f"Số có dãy Collatz dài nhất là: {best_n}")
  print(f"Độ dài dãy tương ứng: {max_len}")
```

```python
#W4A13
def nums(a,b):
  sum_a=0
  sum_b=0
  for i in range(1,a):
    if a%i==0:
      sum_a+=i
  for i in range(1,b):
    if b%i==0:
      sum_b+=i
  if sum_b==a and sum_a==b:
    return True
  else:
    return False
if __name__=="__main__":
  a,b=map(int,input().split())
  if nums(a,b):
    print ("true")
  else:
    print("false")
```
