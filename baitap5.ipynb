```python
import math
def prime(n):
  if n < 2:
    return False
  else:
    for i in range(2,math.isqrt(n)+1):
      if n % i == 0:
        return False
    return True
if __name__=='__main__':
  n=int(input())
  if prime(n):
    print('True')
  else:
    print('False')
```

```python
def perfect_num(n):
  sum=0
  for i in range(1,n):
    if n % i == 0 :
      sum+=i
  if sum == n:
    return True
  else:
    return False
if __name__ == '__main__':
  n=int(input())
  if perfect_num(n):
    print('True')
  else:
    print('False')

```

```python
def factorial(n):
  tich=1
  for i in range(2,n+1):
    tich*=i
  return tich
if __name__=='__main__':
  n=int(input())
  print(factorial(n))
```

```python
def calc(a,b,operat):
  a,b=float(a),float(b)
  if operat == '+':
    return '%.2f' % (a+b)
  elif operat == '-':
    return '%.2f' % (a-b)
  elif operat == '*':
    return '%.2f' % (a*b)
  elif operat == '/':
    return '%.2f' % (a/b)
if __name__=='__main__':
  a,operat,b=map(str,input().split())
  print(calc(a,b,operat))
```

```python
def hamming(a,b):
  a1=[]
  b1=[]
  while a > 0:
    a1.append(str(a%2))
    a//=2
  a1=a1[::-1]
  while b > 0:
    b1.append(str(b%2))
    b//=2
  b1=b1[::-1]
  max_len=max(len(a1),len(b1))
  a1 = ['0'] * (max_len - len(a1)) + a1
  b1 = ['0'] * (max_len - len(b1)) + b1
  dist=0
  for i,j in zip(a1,b1):
    if i != j:
      dist+=1
  return dist
if __name__=='__main__':
  a,b=map(int,input().split())
  print(hamming(a,b))
```

```python
def sum(n):
  tong=0
  while n != 0:
    tong+= n % 10
    n//=10
  return tong
if __name__=='__main__':
  n=int(input())
  print(sum(n))
```

```python
def dang_cau(a,b):
  mapping={}
  mapped_values = set()

  for ca, cb in zip(a, b):
      if ca in mapping:
          if mapping[ca] != cb:
              return False
      else:
          if cb in mapped_values:
              return False
          mapping[ca] = cb
          mapped_values.add(cb)
  return True
if __name__=='__main__':
  a,b=map(str,input().split())
  if dang_cau(a,b):
    print('true')
  else:
    print('false')

```
