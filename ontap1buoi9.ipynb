```python
def top3_frequent_names(names) -> list[str]:
  fre = {}
  for name in names:
    fre[name] = fre.get(name,0) + 1
    sorted_name = sorted(fre.items(),key = lambda x : -x[1])
  result = []
  for i in range(min(3,len(sorted_name))):
    result.append(sorted_name[i][0])
  return result
names = input().split()
print(top3_frequent_names(names))
```

```python
def most_frequent_triple(triples) -> tuple[int,int,int]:
  try:
    fre = {}
    if not triples:
      raise ValueError
    for x in triples:
      fre[x] = fre.get(x,0) + 1
    max_fre = max(fre.values())
    for key,val in fre.items():
      if val == max_fre:
        return key
  except ValueError:
    pass
triples = eval(input())
print(most_frequent_triple(triples))



```

```python
def reconcile_users(registered, paid) -> tuple[list, list, list]:
  registered = set(registered)
  paid = set(paid)
  paid_not_registered = paid.difference(registered)
  registered_not_paid = registered.difference(paid)
  both = registered.intersection(paid)
  return f"({sorted(paid_not_registered)}, {sorted(registered_not_paid)}, {sorted(both)})"
registered = eval(input())
paid = eval(input())
print(reconcile_users(registered,paid))
```

```python
def squares_div_by3_not6(arr) -> list[int]:
  return [x**2 for x in arr if x % 3 == 0 and x % 6 != 0]
arr = list(map(int,input().split()))
print(squares_div_by3_not6(arr))
```

```python
def even_numbers_in_matrix(matrix) -> list[int] :
  lst = []
  for x in matrix:
    lst.extend(x)
  return sorted([x for x in lst if x % 2 == 0])
matrix = []
n = int(input())
for i in range(n):
  matrix.append(list(map(int,input().split())))
print(even_numbers_in_matrix(matrix))
```

```python
def contains_nearby_duplicate(nums, k) -> bool:
  n = len(nums)
  for i in range(n):
    for j in range(i+1,n):
      if nums[i] == nums[j] and (j-i) <= k:
        return True
  return False
nums = list(map(int,input().split()))
k = int(input())
print(contains_nearby_duplicate(nums,k))
```

```python
def merge_similar_items(items1, items2) -> list[list[int]]:
  weight = {}
  for x,y in items1:
    weight[x] = weight.get(x,0) + y
  for x,y in items2:
    weight[x] = weight.get(x,0) + y
  return [[v,weight[v]] for v in sorted(weight)]
items1 = eval(input())
items2 = eval(input())
print(merge_similar_items(items1,items2))


```

```python
def farm_configurations(n) -> list[tuple[int,int]]:
  res = []
  for i in range(n+1):
    cow = i
    chicken = (n - 4* i) / 2
    if abs(chicken - int(chicken)) == 0 and chicken >= 0:
      res.append((int(chicken),cow))
  return res
n = int(input())
print(farm_configurations(n))
```

```python
def can_make_odd_sum(a) -> bool:
  n = len(a)
  odd = [x for x in a if x % 2 != 0]
  if len(odd) == n or len(odd) == 0:
    return True
  else: return False
a = list(map(int,input().split()))
print(can_make_odd_sum(a))
```

```python
def balanced_array(n):
    if (n // 2) % 2 != 0:
        return ("NO", [])

    half = n // 2
    evens = [2 * i for i in range(1, half + 1)]
    odds = [2 * i - 1 for i in range(1, half)]

    last_odd = sum(evens) - sum(odds)
    odds.append(last_odd)

    return ("YES", evens + odds)


n = int(input())
print(balanced_array(n))
```
